```python
import numpy as np
import pandas as pd
from sklearn import datasets
from scipy.stats import skew
import pylab as p
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
df
```

```python
def check_dist_from_data(x):
    if skew(x)>0:
        print("Left tailed")
    elif skew(x)<0:
        print("Right tailed")
    else:
        print("Normal")
        
check_dist_from_data(np.array(df['mean area']))
```

```python
x = df['mean area']

sns.set(rc={'figure.figsize':(8,6)}) # Fig size and labelling
sns.kdeplot(x)
```

```python
x1 = np.random.normal(0, 2, 10000)
check_dist_from_data(x1)
```

```python
sns.set(rc={'figure.figsize':(8,6)}) # Fig size and labelling
sns.kdeplot(x1)
```
