```python
# Import the libraries
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import os
from matplotlib.pyplot import figure
```

```python
path = '../data/Oasis_original_new_renamed/'

original_graphs = [nx.read_gpickle(path + g) for g in os.listdir(path)]
nb_orig = [nx.number_of_nodes(g) for g in original_graphs]
```

```python
path_n = '../data/simu_graph/NEW_SIMUS_JULY_11/4/noise_100,outliers_varied/graphs/'
noise_graphs = [nx.read_gpickle(path_n + g) for g in os.listdir(path_n)]
nb_noise_100 = [nx.number_of_nodes(g) for g in noise_graphs]
```

```python
path_n =  '../data/simu_graph/NEW_SIMUS_JULY_11/4/noise_200,outliers_varied/graphs/'
noise_graphs = [nx.read_gpickle(path_n + g) for g in os.listdir(path_n)]
nb_noise_200 = [nx.number_of_nodes(g) for g in noise_graphs]
```

```python
path_n =  '../data/simu_graph/NEW_SIMUS_JULY_11/4/noise_400,outliers_varied/graphs/'
noise_graphs = [nx.read_gpickle(path_n + g) for g in os.listdir(path_n)]
nb_noise_400 = [nx.number_of_nodes(g) for g in noise_graphs]
```

```python
path_n =  '../data/simu_graph/NEW_SIMUS_JULY_11/4/noise_1000,outliers_varied/graphs/'
noise_graphs = [nx.read_gpickle(path_n + g) for g in os.listdir(path_n)]
nb_noise_1000 = [nx.number_of_nodes(g) for g in noise_graphs]
```

```python
# List of graph folders to plot
graphs = ['original', 'noisy']

    
# Draw the density plot
sns.distplot(nb_orig, hist = False, kde = True,
         kde_kws = {'shade': True,'linewidth': 3}, label=graphs[0])

# Draw the density plot
sns.distplot(nb_noise_100, hist = False, kde = True,
         kde_kws = {'shade': True,'linewidth': 3}, label=graphs[1]+'_100')

# Draw the density plot
sns.distplot(nb_noise_200, hist = False, kde = True,
         kde_kws = {'shade': True,'linewidth': 3}, label=graphs[1]+'_200')

# Draw the density plot
sns.distplot(nb_noise_400, hist = False, kde = True,
         kde_kws = {'shade': True,'linewidth': 3}, label=graphs[1]+'_400')

# Draw the density plot
sns.distplot(nb_noise_1000, hist = False, kde = True,
         kde_kws = {'shade': True,'linewidth': 3}, label=graphs[1]+'_1000')
    

# Plot formatting
plt.legend(prop={'size': 10}, title = 'Graphs')
plt.title('Density Plot')
plt.xlabel('Number of nodes')
plt.ylabel('Density')
plt.show()

#plt.savefig('orig_noise',dpi=200)
```
